```python
# Capping Outliers

# Quantiles
# 25th %iles
# 0
# 25
# 50
# 75
# 100

import numpy as np

MyData = np.random.normal(loc=90,scale=5,size=1000)
```

```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.hist(MyData,bins=50)
```

```python
# UC = Q3 + 1.5 * IQR
# LC = Q1 - 1.5 * IQR

# Q1 - 25% 
# Q3 - 75%
# IQR - Inter Quartile Range  - Q3 - Q1

# Capping
# Any value > UC -> replace them with UC
# Any value < LC -> replace them with LC

MyData = pd.Series(MyData)
MyData.quantile([0,1])

# Quantile takes a list as input
# This list only accepts numbers bn 0 and 1
# where 0 is 0% and 1 is 100%
quadrants = MyData.quantile([0,0.25,0.5,0.75,1])
quadrants

Q1 = quadrants.loc[0.25]
Q3 = quadrants.loc[0.75]
IQR = Q3 - Q1

UC = Q3 + 1.5 * IQR
LC = Q1 - 1.5 * IQR

MyData1 = MyData
MyData1.loc[MyData1 > UC] = UC
MyData1.loc[MyData1 < LC] = LC
plt.boxplot(MyData1)
```

```python
class Account(object):
    def __init__(self, holder = "", number=0, balance=0,credit_line=1500): 
        self.Holder = holder 
        self.Number = number 
        self.Balance = balance
        self.CreditLine = credit_line

    def deposit(self, amount): 
        self.Balance = self.Balance + amount

    def withdraw(self, amount): 
        if(self.Balance - amount < -self.CreditLine):
            # coverage insufficient
            return False  
        else: 
            self.Balance -= amount 
            return True

    def balance(self): 
        return self.Balance

    def transfer(self, target, amount):
        if(self.Balance - amount < -self.CreditLine):
            # coverage insufficient
            return False  
        else: 
            self.Balance -= amount 
            target.Balance += amount 
            return True
```

```python
acc1 = Account(holder="Vipin",number=362473573,balance=3000)
acc2 = Account(holder="Sridhar",number=362377563,balance=4000)
```
