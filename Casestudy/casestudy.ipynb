```python
import pandas as pd
import matplotlib.pyplot as plt
data_path = "./SalesData.csv"
salesdata = pd.read_csv(data_path)
df = pd.DataFrame(salesdata)
```

```python
df1 = df.groupby(["Region"])[["Sales2015","Sales2016"]].sum().add_prefix("Sumof").reset_index()
df1.plot.bar()
```

```python
# pie chart
df1 = df.groupby(["Region"])[["Sales2016"]].sum().add_prefix("Sumof").reset_index()
df1
```

```python
plt.pie(x="SumofSales2016",labels="Region",data = df1,autopct='%d',startangle=90, explode = (0.1,0.1,0.5))
```

```python
from plotnine import *
df2 = df.groupby(["Region","Tier"])[["Sales2015"]].sum().reset_index()
(
    ggplot(data = df2)+
    aes(x = "Tier", y= "Sales2015")+
    geom_bar(stat = "identity",fill = "seagreen")+ 
    facet_grid(".~Region") 
 
)
```
